```python
from PIL import Image
import numpy as np

def resize_image(image_path, target_size):
    # 读取图像
    image = Image.open(image_path)

    # 获取原始图像的尺寸
    width, height = image.size

    # 计算目标图像的缩放比例
    scale = max(target_size[0] / width, target_size[1] / height)

    # 计算调整后的图像大小
    new_width = int(width * scale)
    new_height = int(height * scale)

    # 缩放图像
    resized_image = image.resize((new_width, new_height), Image.ANTIALIAS)

    # 创建目标尺寸的画布
    canvas = Image.new("RGB", target_size, (255, 255, 255))

    # 将调整后的图像粘贴到画布中心
    offset = ((target_size[0] - new_width) // 2, (target_size[1] - new_height) // 2)
    canvas.paste(resized_image, offset)

    return canvas

# 读取图像并调整尺寸
image_path = "/content/plot.png"
target_size = (256, 256)
resized_image = resize_image(image_path, target_size)

# 保存调整尺寸后的图像
output_path = "/content/plot_new.jpg"
resized_image.save(output_path)

print("图像调整尺寸成功。")















import cv2

# 读取图像
image = cv2.imread('/content/plot_new.jpg')

# 转换为灰度图像
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# 保存灰度图像，并保持原始尺寸
cv2.imwrite('/content/plot_finished.png', gray_image, [cv2.IMWRITE_PNG_COMPRESSION, 0])

print("保存成功！")



```
